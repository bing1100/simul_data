```python
import calendar
import numpy as np
import pandas as pd
import scipy as sp
```

```python
census = pd.read_csv("./qid/statcan.csv")
age_col = census["Age (in single years), average age and median age (128)"]
census = census[age_col.map(len) <= 2]
gen_col = census["Gender (3a)"]
census = census[gen_col.map(lambda x: x in ["Men+", "Women+"])]
census = census[["Age (in single years), average age and median age (128)", "Gender (3a)", "VALUE"]]
census = census.rename(columns={"Age (in single years), average age and median age (128)": "age", "Gender (3a)":"gender", "VALUE":"number"})

men_census = census[gen_col == "Men+"]
women_census = census[gen_col == "Women+"]

d = {
    'age' : men_census['age'].values,
    'men' : men_census['number'].values,
    'woman' : women_census['number'].values,
}

print(d)

census = pd.DataFrame(d)

display(census)

census.to_csv("./qid/census.csv", index=False)
```
